### Data Augmentation

```python
import warnings
warnings.simplefilter('ignore')
import cv2
from skimage.io import imshow
```

```python
image = cv2.imread('damaged/1.jpg')
imshow(image)
```

### horizontal flip

```python
horizontal_flip = image[:, ::-1]
imshow(horizontal_flip)
```

### vertical flip

```python
vertical_flip = image[::-1, :]
imshow(vertical_flip)
```

### Rotate image

```python
from skimage.transform import rotate
image_with_rotation = rotate(image, 180)
imshow(image_with_rotation)
```
