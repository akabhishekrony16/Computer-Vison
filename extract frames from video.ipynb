### Extract Frames from Videos 

```python
import cv2

def video_to_frames(video_filename,frames=5):
    """Extract frames from video"""
    n_frames = frames
    cap = cv2.VideoCapture(video_filename)
    video_length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) - 1
    frames = []
    if cap.isOpened() and video_length > 0:
        frame_ids = [0]
        if video_length >= 4:
            for i in range(video_length):
                frame_ids.append(round(video_length * i/(n_frames-1)))
            """
            frame_ids = [0, 
                         round(video_length * 0.1), 
                         round(video_length * 0.2),
                         round(video_length * 0.3),
                         round(video_length * 0.4), 
                         round(video_length * 0.5),
                         round(video_length * 0.6),
                         round(video_length * 0.7), 
                         round(video_length * 0.8),
                         round(video_length * 0.9),
                         video_length - 1]
            """
        count = 0
        success, image = cap.read()
        while success:
            if count in frame_ids:
                frames.append(image)
            success, image = cap.read()
            count += 1
    return frames
```

```python
frames = video_to_frames('big_buck_bunny_720p_5mb.mp4',30)
```
